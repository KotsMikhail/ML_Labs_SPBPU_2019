```python
import numpy as np
```

```python
N = 1000
X = np.linspace(0, 1, N)
eps = np.random.normal(0, 100, N)
a = 7
y = a * X + eps
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, y)
plt.show()
```

```python
k = np.dot(X, y) / np.dot(X, X)
print(k)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
```

```python
X = np.linspace(0, 1, N)
eps = np.random.normal(0, 1, N)
a = 7
y = a * X + eps
X = np.vstack([X, X]).T
X.shape

# r = LinearRegression().fit(X, y)
r = Lasso(alpha=0.001).fit(X, y)

print(r.coef_)
print(r.intercept_)

```

```python
X = np.linspace(0, 1, N)
eps = np.random.normal(0, 1, N)
a = 7
y = a * X + eps
X = np.vstack([X, X]).T
A = np.array([1.0, 1.0])
c = 0.01
rate = 0.9
for i in range(100):
    grad = 2 * np.dot((y - np.dot(X, A)), X)
    A = A + c * grad
    c *= rate
print(A)
print(A.sum())
```
